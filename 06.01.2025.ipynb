```python
# else as an fallback

username = 'besant'

if len(username) > 0:
    print(f"Welcome, {username}!")
else:
    print("Error: Username cannot be empty")
```

```python
# elif statement


age = 22

if age < 13:
    print('child')
elif age < 20:
    print('teenager')
elif age < 65:
    print('adult')
elif age >=65:
    print('senior')
```

```python
# short-hand IF..ELSE

print('b is greater than a') if b>a else print('a is greater than b')
```

```python
# Nested IF condition

age = 22
has_license = True

if age>=18:
    if has_license:
        print('you can drive')
    else:
        print('you need to get a license')
else:
    print('you are too young to drive')
```

## While loop

```python
i = 1

while i <= 6:
    print(i)
    i = i + 1
```

```python
i = 1

while i<=3:
    print(i)
    i+= 1
```

```python
# break statement

i = 1

while i<=10:
    print(i)
    
    if i==3:
        break
        
    i+= 1
```
