```python
# init method/ functions


class person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        
    # self represents the current instance of the class
    
obj = person('Besant','15')
obj2 = person('Raj','20')

print(obj.age, obj.name)
print(obj2.age, obj2.name)
```

```python
class person:
    def __init__(self, name, age):
        self.name = name
        self.age = age 
        
    def greet(self):
        print("Hello, welcome " + self.name)
        
obj1 = person('Monisha', '25')
obj1.greet()
```

```python
class person:
    def __init__(current, name, age):
        current.name = name
        current.age = age 
        
    def greet(now):
        print("Hello, welcome " + now.name)
        
obj1 = person('Monisha', '25')
obj1.greet()
```

```python
# class to check whether the student has passed or failed


# class creation
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks
        
    def result(self):
        if self.marks >= 40:
            return "pass"
        else:
            return "fail"
        
# object creation
s1 = Student('Arun', 65)
s2 = Student('Kalai', 36)
s3 = Student('Niroop', 40)


# Use the objects to find the result
print(s1.name,":", s1.result())
print(s2.name,":", s2.result())
print(s3.name,":", s3.result())
```
