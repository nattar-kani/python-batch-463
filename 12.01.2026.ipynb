```python
import json
```

```python
# convert python to json

x = '{"course" : "Python/ AI","by" : "Besant technologies","mode" : "online"}'

print(x)
print(type(x))

y = json.loads(x)

print(y)
print(type(y))
print(y['mode'])
```

## RegEx

```python
import re
```

```python
txt = "this is my phone number 1234567890"
```

```python
x = re.search(r"\d{10}", txt)

if x:
    print(x.group())
```
