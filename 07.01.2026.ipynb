## For loop

```python
courses = ['python', 'java']

for c in courses:
    print(c)
```

```python
for i in 'besant':
    print(i)
```

```python
# break statement

for i in 'besant':
    print(i)
    
    if i=='a':
        break
```

```python
courses = ['python', 'java', 'c', 'c++']

for c in courses:
    print(c)
    
    if c == 'c':
        break
```

```python
# along with range

for x in range(5):
    print(x)
```

```python
for x in range(11,16):
    print(x)
```

```python
for x in range(0,31,3):
    print(x)
```

```python
# nested loops
courses = ['python', 'java', 'c++']
hrs = [15,20,30]

for c in courses:
    for h in hrs:
        print(c,h)
```

```python
# print first 10 natural numbers using for loop

for i in range(1,11):
    print(i)
```

```python
# print all the even numbers in the range 11-20
for i in range(11,21):
    if i%2==0:
        print(i)
```

```python
# print to calculate the sum of all numbers from 1 to a given number
# range (1-6) 1+2+3+4+5 = 15

total = 0

for i in range(1,6):
    total+=i
print(total)
```

```python
# to check if the string is a palindrome

string = 'madam'
rev_str = ''

for i in string:
    rev_str = i + rev_str
    
if (string == rev_str):
    print('it is a palindrome')
else:
    print('it is not a palindrome')
```

```python
# print the factorial of the given number

num = 6
factorial = 1

for i in range(1,num+1):
    factorial = factorial*i
    
print(factorial)
```
