```python
# parameter vs argument

def add(n1,n2): #n1,n2 are parameters
    print(n1+n2)
```

```python
add(10,4) #10, 4 are arguments
```

```python
# arbituary arguments *args
# accepts any number of positional arguments

def add(*num):
    print(num)
```

```python
add(10,2,3)
```

```python
add(10)
```

```python
# using *args with regular argument

def welcome(greeting, *names):
    for n in names:
        print(greeting,names)
        
welcome("Hello","Ram","Gokul","Shiva")
```

```python
# **kwargs - any number of keyword arguments can be given

def details(**value):
    print("Name:", value["name"])
    print("Age:", value["age"])
    print("Course:", value["course"])
    print("all:", value)
    print("Type of the value:", type(value))
```

```python
details(name = "Monisha",age="24",course="Engineering")
```

## Scope

```python

def myfunc():
    x = 1
    print(x)
    
myfunc()
```

```python
# Local scope

def myfunc():
    x = 1 #local variable
    def innerfunc():
        print(x+5) # local variable
    innerfunc()
    
myfunc()
```

```python
def myfunc():
    y = 1
    print(y)
    
print(y)
myfunc()

```

```python
# global scope

x = 1 #global variable


def myfunc():
    print(x)
    
myfunc()

print(x)
```

```python
# to give global scope for local variable

def myfunc():
    global x
    x = 5
    
myfunc()

print(x)
```

## Decorators

```python
@changecase
def myfunc():
    return "Besant Technologies"

def changecase(func):
    def inner():
        return func().upper()
    return inner

print(myfunc())
```

```python
@changecase
def myfunc():
    return "Besant Technologies"

@changecase
def myfunc2():
    return "Welcome students"

def changecase(func):
    def inner():
        return func().upper()
    return inner

print(myfunc())
print(myfunc2())
```
