## Set

- unordered
- unchangeable
- no duplicates
- heterogeneous collection of data is allowed

```python
s = {'python','java','c++'}
print(s)
print(type(s))
```

```python
s1 = {10,10} # duplicates will not be considered
print(s1)
print(type(s1))
```

```python
s1 = {'python',10, 1, True}
print(type(s1))
```

```python
print(s1) 
# it considers 1 and True as the same value
```

```python
s1 = {'python',10, False, 0}
s1
# it is considering False and 0 as same values
```

```python
len(s)
```

```python
# set() constructor
s1 = set(('python',10, False))
s1
```

```python
# access the item

print(s)

for i in s:
    print(i)
```

```python
# adding items to the set

s.add('testing')
print(s)
```

```python
s1
```

```python
s2 = {55, 'c++', 4.3}
```

```python
s1.update(s2) # we are merging set s2 with set s1
print(s1)
```

```python
lst= [6,7,8]
```

```python
s1.update(lst) # we are merging list lst to set s1
print(s1)
```

```python
# remove an item

s1.remove('python')
print(s1)
```

```python
s1.remove(1)
print(s1)
```

```python
s1.discard(1)
# if it exists, it discards the value else it shows all the items in the set as such
# it wont throw any error, if the item does not exist in the set
print(s1)
```

```python
s1.pop()
s1
```

```python
s2.clear() # to remove all the items from the set
s2
```

```python
type(s2)
```

```python
del s2
s2
```

### join operations

```python
set1 = {'a','b','c'}
set2 = {1,2,3, 'a'}
```

```python
set3 = set1.union(set2) #union operator |
set3
```

```python
set1.union(set2)
```

```python
set1
```

```python
set1 | set2
```

```python
set3 = {'sun','moon'}
```

```python
set1 | set2 | set3
```

```python
# union of set and tuple

tup = ('python','coding')

x = set1.union(tup)
print(x)
```

```python
# update()

set1.update(set2)
set1
```

Both union( ) and update( ) will discard duplicate values

```python
set3 = {3,6,9,15}
set4 = {5,10,15,20}
```

```python
# intersection() or & operator

result = set3.intersection(set4)
result
```

```python
set3.intersection_update(set4)
set3
```

```python
set4
```

```python
set3 = {3,6,9,15}
set4 = {5,10,15,20}
```

```python
# difference() or - operator

result = set3.difference(set4)
result
```

```python
result = set4.difference(set3)
result
```

```python
print(set3)
set3.difference_update(set4)
print(set3)
```

```python
set3 = {3,6,9,15}
set4 = {5,10,15,20}
```

```python
# symmetric difference or ^ operator
result = set3.symmetric_difference(set4)
result
```

## Frozen sets

- immutable version of set
- unique
- unordered
- unchangeable (cannot remove or add items also)

```python
a = frozenset({10,20,30})
print(a)
print(type(a))
```
