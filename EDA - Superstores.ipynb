## Correlation

```python
correlation = df.corr()
correlation
```

```python
sns.heatmap(correlation,cmap='coolwarm',annot=True)
```

## Box plot

```python
NumData = df.select_dtypes(include=[np.number])
NumData
```

```python
sns.boxplot(x='variable',y='value',data=pd.melt(NumData),width=1)
```
