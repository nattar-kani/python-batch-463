## Numpy - continuation

```python
import numpy as np
```

```python
arr = np.array([1,2,3,4,5,6])

# splitting numpy arrays

x = np.array_split(arr,3)

x
```

```python
arr = np.array([[1,2,3,4,5,6],[1,2,3,4,5,6]])

# splitting numpy arrays

x = np.array_split(arr,2)

x
```

```python
# searching in the array for a certain value

arr = np.array([1,2,3,4,5,6])

x = np.where(arr == 3) # returns the index of that particular value
x
```

```python
# searching in the array for a certain value

arr = np.array([1,2,3,4,5,6,4,3,3,4,5,5,5,5])

x = np.where(arr == 5) # returns the index of that particular value
x
```

```python
# searching in the array for a certain value

arr = np.array([1,2,3,4,5,6,4,3,3,4,5,5,5,5])

x = np.where(arr%2 == 0) # returns the index of that particular value
x
```

```python
# sorting arrays

arr = np.array([12,44,22,1])

print(np.sort(arr))
```

```python
arr = np.array([[12,44,22,1],[2,80,11,36]])

print(np.sort(arr))
```

```python
# filtering the values in the array

arr = np.array([12,44,22,1,85,36])

filt = arr > 40

print(arr)

x = arr[filt]

print(x)
```

## Pandas

```python
import pandas as pd
```

```python
dt = {
    'course': ["Python", "AI/ML", "Testing"],
    'TotalHrs': [25,30,30]
}

print(dt)
print(type(dt))
```

```python
df = pd.DataFrame(dt)

print(df)
```

```python
print(type(df))
```

```python
# series is more like a column in the table

x = [1,2,3]

ser = pd.Series(x)

print(ser)
print(type(ser))
```

```python
# series is more like a column in the table

x = [1,2,3]

ser = pd.Series(x, index=['a','b','c'])


print(ser)
print(type(ser))
```

```python
print(ser['b'])
```

```python
# locating a row in a dataframe

dt = {
    'course': ["Python", "AI/ML", "Testing"],
    'TotalHrs': [25,30,30]
}

df = pd.DataFrame(dt)

print(df)
print(type(df))
```

```python
print(df.loc[1])
```

```python
print(df.loc[0:1])
```

```python
df = pd.DataFrame(dt, index = ['course_1','course_2','course_3'])

print(df)
```

```python
print(df.loc['course_2'])
```
