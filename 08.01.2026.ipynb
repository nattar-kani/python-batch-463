```python
# to convert temperature from fahrenheit to celsius

temp_f = 77

temp_c = (temp_f - 32) * 5/9
print(temp_c)
```

```python
# to convert temperature from fahrenheit to celsius

temp_f = 82

temp_c = (temp_f - 32) * 5/9
print(temp_c)
```

```python
def fahrenheit_to_celcius(temp_f):
    return (temp_f - 32) * 5/9


print(fahrenheit_to_celcius(92))
print(fahrenheit_to_celcius(100))
```

```python
# Positional arguments

def details(course, hours):
    print("This is ", course, " course")
    print(course, " course takes ", hours, " hours of session")

details('python', 20)
```

```python

def details(course, hours):
    print("This is ", course, " course")
    print(course, " course takes ", hours, " hours of session")

details(20, 'python')
```

```python
# Keyword arguments
# in short - kwargs

def details(course, hours):
    print("This is ", course, " course")
    print(course, " course takes ", hours, " hours of session")

details(hours = 20, course = 'python')
```

```python
def details(course, hours):
    print("This is ", course, " course")
    print(course, " course takes ", hours, " hours of session")

details(hours = 20, 'python')
```

```python
# positional arg followed by keyword arg 


def details(course, hours):
    print("This is ", course, " course")
    print(course, " course takes ", hours, " hours of session")

details( 'python', hours=20)
```

```python
# for position-only arguments is allowed in the function

def welcome(name, /):
    print("hello ", name)
    
welcome("raj")
welcome("")
welcome(name = "ram")
```

```python
def welcome(name): # it will accept the keyword argument (kwargs)
    print("hello ", name)
    
welcome("raj")
welcome("")
welcome(name = "ram")
```

```python
# for keyword-only arguments is allowed in the function

def welcome(*,name):
    print("hello ", name)
    
welcome(name = "ram")
welcome("raj")
```
