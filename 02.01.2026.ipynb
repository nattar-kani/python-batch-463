```python
# dict() constructor

thisdict = dict(course="ai/ml", mode="online")
print(thisdict)
print(type(thisdict))

# use = instead of colon :, and write key directly 
```

```python
# accessing items using get() method

dictt = {
    "course": "python",
    "hours": 24,
    "mode": "online"
}

dictt.get("hours")
```

```python
dictt.keys() # to access all the keys
```

```python
dictt.values() # to access all the values/ items in the dict
```

```python
# adding new pair to the original dict

x = dictt.keys()

print(x)

dictt["given by"] = "besant technologies"

print(x)
```

```python
# update the existing value in the dict

print(dictt)

y = dictt.values()

print(y)

dictt["hours"] = 20

print(y)
```

```python
dictt
```

```python
dictt.update({"course": "python/ai"}) # using update() method
```

```python
dictt
```

```python
# removing the items using pop()

print(dictt)

dictt.pop("given by")

print(dictt)
```

```python
# removing the item

print(dictt)

dictt.popitem()

print(dictt)
```

```python
# using del

del dictt["hours"]

dictt
```

```python
dictt = {
    "course": "python",
    "hours": 24,
    "mode": "online"
}

# loop to display the key inside the dict

for x in dictt:
    print(x)
    
for x in dictt.keys():
    print(x)
```

```python
# loop to display the values inside the dict

for x in dictt:
    print(dictt[x])
    
for x in dictt.values():
    print(x)
```

```python
# loop to display both key-value pairs

for x,y in dictt.items():
    print(x,y)
```

```python
# copy the dict

mydict = dictt.copy()

mydict
```

```python
mydict1 = dict(dictt)

mydict1
```
