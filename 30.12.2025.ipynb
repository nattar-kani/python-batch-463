## List (continuation)

```python
courses = ['python', 'java', 'c++']
```

```python
courses
```

```python
# loop thro a list

for x in courses:
    print(x)
```

```python
# using index

for x in range(len(courses)):
    print(courses[x])
```

```python
# using while loop

i = 0
while i < len(courses):
    print(courses[i])
    i= i+1
```

```python
# list comprehension
# for x in courses:
#   print(x)

[print(x) for x in courses]
```

```python
# sort - ascending

courses.sort()
print(courses)
```

```python
# sort - descending

courses.sort(reverse= True)
print(courses)
```

```python
# copy()
courses = ['python', 'java', 'c++']

lst = courses.copy()
lst
```

```python
# using list constructor

lst_1 = list(courses)
lst_1
```

```python
hrs = [24,30,18]
hrs
```

```python
# to merge the list using + operator

plan = courses + hrs
plan
```

```python
# extend()

courses.extend(hrs)
courses
```

```python
courses = ['python', 'java', 'c++']
hrs = [24,30,18]

for x in hrs:
    courses.append(x)
    
print(courses)
```

## Tuples

- immutable
- enclosed within ( )
- ordered collection
- indexing works
- heterogeneous data types can be stored in the tuple
- allows duplicates

```python
courses = ('python','c++','java')
print(courses)
print(type(courses))
```

```python
courses = ('python','c++','java', 'java')
print(courses)
print(type(courses))
```

```python
len(courses)
```

```python
# to create a tuple with one item

tuple_1 = ("python")
print(type(tuple_1))
```

```python
tuple_1 = (10)
print(type(tuple_1))
```

```python

tuple_1 = ("python",)
print(type(tuple_1))
```

```python
tuple_1 = (10,)
print(type(tuple_1))
```

```python
tuple_1 = ('python', 17, 5.2, True)
print(type(tuple_1))
```

```python
# tuple() constructor

tup = tuple(('python','java'))
type(tup)
```
